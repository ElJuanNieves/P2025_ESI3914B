#### Problem Statement ####
Create a program that uses a dictionary to map string commands to functions. Allow the user to input a command and execute the corresponding function (e.g., "greet", "square", "cube").



```python
class CommandExecuter:

    def __init__(self):
        self.commands = {}

    def add_command(self, func):
        self.commands[func.__name__] = func
    
    def execute(self, name, foo):
        if name in self.commands:
            return self.commands[name](foo)
        return "function not implemented"
        
def greet(name):
    return f"Hello {name}"

def square(num):
    return num * num

def cube(num):
    return num * num * num

my_executer = CommandExecuter()

my_executer.add_command(greet)
my_executer.add_command(square)
my_executer.add_command(cube)

print(my_executer.execute('greet', 'Jose'))
print(my_executer.execute('square', 2))
print(my_executer.execute('cube', 3))

```
