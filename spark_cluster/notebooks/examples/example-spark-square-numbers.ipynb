#### Creacion de la conexión con el cluster de spark


```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("SquareNumbers") \
    .master("spark://078b2e28e517:7077") \
    .config("spark.ui.port","4040") \
    .getOrCreate()

sc = spark.sparkContext
print("Spark context created successfully")
```

**Create the RDD**

```python
import random
sc = spark.sparkContext
data = [random.randint(0, 517) for i in range(500)]

rdd = sc.parallelize(data)
rdd
```

**Using a distributed approach to compute squares of numbers**

```python
rdd_squares = rdd.map(lambda x: x ** 2)
```

**Collect results to a python collection to explore them**

```python
squares_lst = rdd_squares.collect()
squares_lst
```

**Close Spark Session**

```python
sc.stop()
```
